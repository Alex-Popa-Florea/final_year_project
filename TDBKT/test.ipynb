```python
from mastery import SkillMastery, ContributionSkillMastery, GroupSkillMastery
import numpy as np
from matplotlib import pyplot as plt
```

```python
def plot_mastery(mastery, title):
    plt.figure(figsize=(10, 2))
    plt.plot(np.arange(mastery.t), np.array(mastery.H), linestyle='-')

    plt.xlabel('Time Step')
    plt.ylabel('Mastery Probability')
    plt.title(title)

    plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1])
    plt.show()

def plot_observation(mastery):
    plt.figure(figsize=(10, 2))
    plt.plot(np.arange(mastery.t), np.array(mastery.o), linestyle='-')

    plt.xlabel('Time Step')
    plt.ylabel('Observation')

    plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1])
    plt.show()

def plot_contribution(mastery):
    plt.figure(figsize=(10, 2))
    plt.plot(np.arange(mastery.t), np.array(mastery.c), linestyle='-')

    plt.xlabel('Time Step')
    plt.ylabel('Contribution')

    plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1])
    plt.show()
```

```python
student_1 = SkillMastery(p_L_0=0.5, p_S=0.1, p_G=0.1, E_k=60, n=10)
student_2 = ContributionSkillMastery(p_L_0=0.5, p_S=0.1, p_G=0.1, E_k=60, n=10)
student_3 = ContributionSkillMastery(p_L_0=0.5, p_S=0.1, p_G=0.1, E_k=60, n=10)
group = GroupSkillMastery(2, p_L_0=0.5, p_S=0.1, p_G=0.1, p_T=0.5, p_F=0.3, n=10, discussion_time=40, solve_time=20)

for i in range(80):
    if 22 <= i < 30:
        student_1.step(True)
        student_2.step(0.5)
        student_3.step(0.5)
        group.step(1, [1, 1])
    elif 45 <= i:
        student_1.step(True)
        x = np.random.rand()
        student_2.step(0.3)
        student_3.step(1)
        group.step(1, [0.5, 1])
    else:
        student_1.step(False)
        student_2.step(0)
        student_3.step(0)
        group.step(o_t=0, c_ts=[0, 0])
```

```python
plot_observation(student_1)
plot_mastery(student_1, "TD-BKT")
plot_contribution(student_2)
plot_mastery(student_2, "CTD-BKT")
print(student_2)
plot_contribution(student_3)
plot_mastery(student_3, "CTD-BKT")
print(student_3)
for student in group.students:
    plot_contribution(student)
    plot_mastery(student, "CLTD-BKT")
    print(student)
```
